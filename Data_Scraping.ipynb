```python
import requests
from bs4 import BeautifulSoup
from urllib.request import unquote
import PyPDF2
from itertools import chain
```

```python
import pandas as pd
```

```python
#Create text and pdf folders
os.chdir(os.getcwd())
if not os.path.exists("text"):
    os.makedirs("text")
if not os.path.exists("pdf"):
    os.makedirs("pdf")
```

```python
#Source: https://github.com/maksimKorzh/one-time-scrapers/blob/master/scrapers/scrape_pdf/scrape.py

url = 'http://laserowatwierdza.pl/wyniki/'
lists = []
# make HTTP GET request to the target URL
print('HTTP GET: %s', url)
response = requests.get(url)

# parse content
content = BeautifulSoup(response.text, 'lxml')

# extract URLs referencing PDF documents
all_urls = content.find_all('a')

# loop over all URLs
for url in all_urls:
    # try URLs containing 'href' attribute
    try:
        # pick up only those URLs containing 'pdf'
        # within 'href' attribute
        if 'pdf' in url['href']:
            # init PDF url
            pdf_url = ''
            
            # append base URL if no 'https' available in URL
            if 'https' not in url['href']:
                pdf_url = 'https://dyysg.org.uk' + url['href']

            # otherwise use bare URL
            else:
                pdf_url = url['href']
            
            # make HTTP GET request to fetch PDF bytes
            print('HTTP GET: %s', pdf_url)          
            pdf_response = requests.get(pdf_url)
            # extract  PDF file name
            filename = unquote(pdf_response.url).split('/')[-1].replace(' ', '_')
            
            
            # write PDF to local file
            with open('./pdf/' + filename, 'wb') as f:
                lists.append(filename)
                # write PDF to local file
                #f = PyPDF2.PdfFileReader(f) 
                f.write(pdf_response.content)
                
    # skip all the other URLs
    except:
        pass
    
```

```python
#to txt
for i in range(0,len(lists)):
    pdfFileObj = open('./pdf/' + lists[i], 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    #print(pdfReader.numPages)
    pageObj = pdfReader.getPage(2)
    text = pageObj.extractText()
    with open('./text/' + lists[i] + '.txt', 'w') as f2:
        f2.write(text)
    pdfFileObj.close() 
```

```python
#extracting stats from second page
def stats_fun(page):
    
    pdfReader = PyPDF2.PdfFileReader(page)
    pageObj = pdfReader.getPage(2)
    txt = pageObj.extractText()
    stats = []
    for i in range(1,int(len(txt.split("\n")[17:-1])/10) + 1):
        stats.append(txt.split("\n")[17:-1][-5 + 10*i :10*i])
        #print((len(txt.split("\n")[17:-1])/10) + 1)
    return stats
```

```python
#extracting injuries for pages 3:
def injuries2(pdf):
    pdfReader = PyPDF2.PdfFileReader(pdf)
    #pageObj = pdfReader.getPage(3)
    stats = [] 
    for k in range(3,pdfReader.numPages):
        #print(pdfReader.numPages)
        #licznik = licznik + 1
        #print(licznik)
        pageObj = pdfReader.getPage(k)
        txt = pageObj.extract_text()             
        stats.append(txt.split("get",1)[1].split("\n")[1])
    return stats
```

```python
#recall functions on all scrapped pdfs
stats = []
inj = []
for element in lists:
    pdfFileObj = open('./pdf/' + element, 'rb')
    stats2 = stats_fun(pdfFileObj)
    inj2 = injuries2(pdfFileObj)
    inj.append(inj2)
    stats = stats2 + stats # list conection
    
```

```python
#reverse inj list
inj.reverse()
#inj
```

```python
#reshape inj to 1 dim list
inj_flat = list(chain.from_iterable(inj))
```
